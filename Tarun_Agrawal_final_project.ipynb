### Overall Social Media Platforms usage:

The graph below shows us the social media platforms in their order of usage. The table has been created to find the top 10 social media platforms.

```python
# Finding the unique social media platforms:
social_media_platforms = data2["social Media Platforms"].str.lower().str.replace(" ", "").str.split(pat = ";|,", expand = True).stack().unique()
social_media_platforms

# Counting for each platform:
df_socialmedia = pd.DataFrame(index = social_media_platforms)
counts = []
for i in social_media_platforms:
    platforms_count = 100 * (data2["social Media Platforms"].str.contains(i, case = False).mean())
    counts.append(platforms_count)

df_socialmedia["counts"] = counts
df_socialmedia.reset_index()

# Sorting by count and creating a bar chart:
df_socialmedia = df_socialmedia.sort_values(by = "counts", ascending = False)
df_socialmedia.plot(kind = 'bar', title = "Percent of users using the Social Media Platform", legend = None)
# plt.savefig("figure2.png", bbox_inches = 'tight') # - This is to save this figure to the current working directory.
```

```python
top_platforms = df_socialmedia.head(10)
print(top_platforms)
```

Around 98.35 percent of users used Whatsapp, followed by facebook (91.52 percent) and instagram(80.70 percent). These are the 3 most used social media platforms.

### Social Media Usage by gender:

This part shows the social media usage by gender.

```python
# Slicing the Gender, and social Media Platforms column for the purpose of analysis:
social_media_gender = data2[['Gender','social Media Platforms']]
social_media_gender.head()
```

```python
# Create a list using str.split,and then use explode function to get each value into row
social_media_gender["platforms"] = social_media_gender['social Media Platforms'].str.lower().str.replace(" ", "").str.split(";|,")
social_media_gender = social_media_gender.explode('platforms')[["Gender","platforms"]]

# Finding the platform usage by gender, and creating a pie chart for the same:
social_media_gender_pivot = social_media_gender.pivot_table(index = ["platforms"], columns = "Gender" , aggfunc = 'size', sort = True)
social_media_gender_pivot.fillna(0)

# Finding top platforms by male and female.
top_platform_female = social_media_gender_pivot.sort_values(by = "Female", ascending = False).head(10)
top_platform_male = social_media_gender_pivot.sort_values(by = "Male", ascending = False).head(10)

# Creating pie charts:
top_platform_female.plot.pie(y = "Female", figsize = (12,10), autopct ='%1.1f%%', legend = None, ylabel = "", title = "Female - Top 10 Platforms")
# plt.savefig("figure3.png",bbox_inches = 'tight') # - This is to save this figure to the current working directory.

top_platform_male.plot.pie(y = "Male", figsize = (12,10), autopct ='%1.1f%%', legend = None, ylabel = "", title = "Male - Top 10 Platforms")
# plt.savefig("figure4.png",bbox_inches = 'tight') # - This is to save this figure to the current working directory.
```

The top 3 platforms amongst males and females are facebook, instagram and whatsapp. Twitter is the 4th most used platform amongst males, while snapchat is the 4th amongst females. Other than that, Tiktok is in the top 10 used platforms for females but not for males, and googleplus is in the 10 most used platforms for males, but not females.

# Predictive Analysis:

The outcome variable was changed from Strongly Agree, Agree, Neutral, Disagree and Strongly Disagree to Binary Outcomes, Agree (Strongly Agree, Agree) and Disagree( Neutral, Disagree and Strongly Disagree). 

Post this 4 classification models were implemented to predict the willingness of students to learn online:
1. Logistic Regression
2. Decision Tree
3. K Neighbour Classification
4. Linear Discriminent Analysis

### Logistic Regression :

```python
### Data Slicing for Regression:
data_regression = data2.drop(['Education Level/Certificates (You can choose more than one)',
                              'social Media Platforms', 'emerging technologies',
                              'How do you access your social media platforms? (You can choose more than one)',
                              'Why do you use social networks? (You can choose more than one)',
                              'Are you presently involved in e-learning using any social media platform or emerging technologies earlier highlighted?'],
                            axis = 1)

# Converting output varible to binary: 
values = ['Agree', 'Strongly Agree', 'Neutral', 'Disagree','Strongly Disagree']#
replace = ['Agree','Agree','Disagree','Disagree','Disagree']
data_regression['I will be willing to use social media platforms and emerging technologies for learning'] = data_regression['I will be willing to use social media platforms and emerging technologies for learning'].replace(values, replace)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
data_regression['I will be willing to use social media platforms and emerging technologies for learning'] = le.fit_transform(data_regression['I will be willing to use social media platforms and emerging technologies for learning'])
```

```python
X1 = data_regression.drop(['I will be willing to use social media platforms and emerging technologies for learning'], axis = 1)
Y1 = data_regression['I will be willing to use social media platforms and emerging technologies for learning']
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(sparse = False)
X = ohe.fit_transform(X1)
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,Y1, test_size=0.30, random_state = 24)
```

```python
# Hypertuning:
from sklearn.model_selection import GridSearchCV
c_space = np.linspace(-5, 8, 15)
param_lr = {"penalty" : ['l1', 'l2', 'elasticnet', 'none'],
            "C": c_space}
lr_cv = GridSearchCV(lr, param_lr, cv = 5)
lr_cv.fit(X_train, y_train)
lr_cv.best_score_
lr_cv.best_params_
```

```python
lrmodel = LogisticRegression(C = 0.5714285714285712, penalty = 'l2')
lrmodel.fit(X_train,y_train)
```

```python
lrmodel.predict(X_test)
```

```python
lrmodel.score(X_test,y_test)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
```

```python
cm = confusion_matrix(y_test, lrmodel.predict(X_test))
ConfusionMatrixDisplay(cm, display_labels=['Agree:0','Disagree:1']).plot()
plt.savefig("figure12",bbox_inches = 'tight')
```

```python
print(classification_report(y_test, lrmodel.predict(X_test)))
```

### Decision Tree Algorithm:

```python
from sklearn.tree import DecisionTreeClassifier
tree = DecisionTreeClassifier()
```

```python
# Data Slicing for Decision Trees:
data_dt = data2.drop(['Education Level/Certificates (You can choose more than one)',
                              'social Media Platforms', 'emerging technologies',
                              'How do you access your social media platforms? (You can choose more than one)',
                              'Why do you use social networks? (You can choose more than one)',
                              'Are you presently involved in e-learning using any social media platform or emerging technologies earlier highlighted?'],
                            axis = 1)
```

```python
values = ['Agree', 'Strongly Agree', 'Neutral', 'Disagree','Strongly Disagree']#
replace = ['Agree','Agree','Disagree','Disagree','Disagree']
data_dt['I will be willing to use social media platforms and emerging technologies for learning'] = data_dt['I will be willing to use social media platforms and emerging technologies for learning'].replace(values, replace)
```

```python
X1 = data_dt.drop(['I will be willing to use social media platforms and emerging technologies for learning'], axis = 1)
Y1 = data_dt['I will be willing to use social media platforms and emerging technologies for learning']
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(sparse = False)
X = ohe.fit_transform(X1)
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,Y1, test_size=0.30, random_state = 24)
y_test
```

```python
# Hyper Paramter Tuning For Decision trees:
from scipy.stats import randint
# Setup parameters:
param_dt = {"criterion":["gini","entropy"]}
tree_cv = GridSearchCV(tree, param_dt , cv= 5)
tree_cv.fit(X_train , y_train)
tree_cv.best_params_
tree_cv.best_score_
```

```python
dtmodel = DecisionTreeClassifier(criterion = 'gini', random_state=0)
dtmodel.fit(X_train, y_train)

y_pred1 = dtmodel.predict(X_test)

print("Accuracy Score:", accuracy_score(y_test, y_pred1))
```

```python
cm2 = confusion_matrix(y_test, y_pred1)
ConfusionMatrixDisplay(cm2, display_labels=['Agree','Not Agree']).plot()
plt.savefig("figure13",bbox_inches = 'tight')
print(classification_report(y_test, y_pred1))
```

### K Neighbour Classification:

```python
# Hypertuning parameters for K Neighbour:
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
param_knn = {"n_neighbors":np.arange(1,10)}
knn_cv = GridSearchCV(knn, param_knn , cv= 5)
knn_cv.fit(X_train , y_train)
knn_cv.best_score_
knn_cv.best_params_
```

```python
from sklearn.neighbors import KNeighborsClassifier
model2 = KNeighborsClassifier(n_neighbors = 6)
model2.fit(X_train, y_train)

y_pred2 = model2.predict(X_test)

print("Accuracy Score:", accuracy_score(y_test, y_pred2))
```

```python
cm3 = confusion_matrix(y_test, y_pred2)
ConfusionMatrixDisplay(cm3, display_labels=['Agree','Disagree']).plot()
plt.savefig("figure14",bbox_inches = 'tight')
print(classification_report(y_test, y_pred2))
```

### Linear Discriminent Analysis:

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
lda = LDA()
```

```python
param_lda = {"n_components":np.arange(1,10)}
lda_cv = GridSearchCV(lda, param_lda , cv= 5)
lda_cv.fit(X_train , y_train)
lda_cv.best_score_
lda_cv.best_params_
```

```python
model_lda = LDA(n_components = 1)
model_lda.fit(X_train, y_train)

y_pred_lda = model_lda.predict(X_test)

print("Accuracy Score:", accuracy_score(y_test, y_pred_lda))
```

```python
cm4 = confusion_matrix(y_test, y_pred_lda)
ConfusionMatrixDisplay(cm4, display_labels=['Agree','Disagree']).plot()
plt.savefig("figure15",bbox_inches = 'tight')
print(classification_report(y_test, y_pred_lda))
```
